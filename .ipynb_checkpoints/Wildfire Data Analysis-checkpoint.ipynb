## Histograms everything

Which state is the most fire-prone?

```python
state_df = pd.read_sql('SELECT STATE FROM fires',conn)
state_df.value_counts('STATE').plot.bar(figsize = (20,10))
plt.xlabel('State')
plt.ylabel('# fires total')
```

Which state has suffered the most acres burned?

```python
state_fireSize_df = pd.read_sql('SELECT STATE, FIRE_SIZE FROM fires',conn)
state_fireSize_df.groupby('STATE')['FIRE_SIZE'].sum().plot.bar(figsize=(20,10))
plt.xlabel('State')
plt.ylabel('Total Acres Burned')
```

```python
year_df = pd.read_sql('SELECT FIRE_YEAR FROM fires ORDER BY FIRE_YEAR ASC',conn)
year_df.value_counts('FIRE_YEAR').sort_index().plot.bar(figsize = (20,10))
plt.xlabel('Year')
plt.ylabel('# Fires')
```

In California, the state with the most wildfires, which counties are the most fire prone? 

```python
county_CA_df = pd.read_sql('SELECT COUNTY FROM fires WHERE STATE IN ("CA")',conn)
county_CA_df["COUNTY"] = county_CA_df["COUNTY"].str.lower()
county_counts = county_CA_df.value_counts('COUNTY',normalize=True)
threshold = 0.005
mask = county_counts > threshold
county_counts = county_counts.loc[mask]
county_counts.plot.bar(figsize = (20,10))
plt.xlabel('Counties in CA')
plt.ylabel('# Fires normalized')
```

```python
county_GA_df = pd.read_sql('SELECT COUNTY FROM fires WHERE STATE IN ("GA")',conn)
county_GA_df["COUNTY"] = county_GA_df["COUNTY"].str.strip()
county_counts = county_GA_df.value_counts('COUNTY',normalize=True)
threshold = 0.007
mask = county_counts > threshold
county_counts = county_counts.loc[mask]
county_counts.plot.bar(figsize = (20,10))
plt.xlabel('Counties in GA')
plt.ylabel('# Fires normalized')
```

Is there a correlation between time of year and number of fires? 

```python
day_df = pd.read_sql('SELECT DISCOVERY_DOY FROM fires',conn)
day_df['DISCOVERY_DOY'].value_counts(normalize=True,sort = False, bins = 12).plot.bar(figsize = (20,10))
plt.xlabel('Day of Year Discovered')
plt.ylabel('# fires (normalized)')
```
